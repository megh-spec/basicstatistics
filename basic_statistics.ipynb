1. What is the difference between descriptive statistics and inferential
statistics? Explain with examples.
   - Descriptive statistics summarize and describe data (mean, median, charts), while inferential statistics use sample data to make predictions or generalizations about a larger population (hypothesis testing, regression). The core difference is scope: descriptive focuses on what is in the current data, whereas inferential looks at what might be in the broader group. For example, calculating the average test score for one class (descriptive) versus using that class's score to predict all students' scores in the state (inferential).

Descriptive Statistics

Purpose: To organize, summarize, and present data in a meaningful way for a specific dataset.

Focus: The sample data itself; no generalizations beyond the data studied.
Tools: Mean, median, mode, range, standard deviation, frequency tables, histograms, bar charts.

Example: Reporting the average age, gender breakdown, and highest sales figures for 100 surveyed customers.

Inferential Statistics

Purpose: To draw conclusions, make predictions, or test hypotheses about a population based on a sample.

Focus: Generalizing findings from a sample to a larger population.
Tools: Hypothesis testing (t-tests, ANOVA), regression analysis, confidence intervals.

Example: Using the survey data from 100 customers to estimate the buying preferences of all 10,000 customers of the company.

2.  What is sampling in statistics? Explain the differences between random
and stratified sampling.
    - Sampling in Statistics

Sampling in statistics is the process of selecting a smaller group of individuals or observations (called a sample) from a larger group (called a population) in order to study and draw conclusions about the entire population. Sampling is used because studying the whole population is often time-consuming, costly, or impractical. A well-chosen sample helps researchers make accurate and reliable inferences about the population.

Random Sampling

Random sampling is a method in which every individual or element of the population has an equal and independent chance of being selected for the sample. Selection is done purely by chance, such as using a lottery method or random number tables.

In random sampling, bias is minimized because no particular group is intentionally favored. It is simple to understand and easy to use when the population is small and homogeneous. However, if the population is large or diverse, random sampling may fail to represent all important subgroups adequately.

Stratified Sampling

Stratified sampling is a method in which the population is first divided into smaller, non-overlapping groups called strata based on specific characteristics such as age, gender, income, or education level. After forming the strata, random samples are taken from each stratum, usually in proportion to their size in the population.

This method ensures that all important subgroups of the population are properly represented in the sample. Stratified sampling generally provides more accurate and reliable results than simple random sampling, especially when the population is heterogeneous. However, it requires more information about the population and is more complex to design.

Difference Between Random Sampling and Stratified Sampling

In random sampling, the entire population is treated as one group, and samples are selected randomly without considering subgroups. In contrast, stratified sampling divides the population into strata and then selects samples from each stratum.

Random sampling is easier to conduct but may not represent all subgroups well. Stratified sampling is more systematic and ensures better representation, but it is more time-consuming and requires prior knowledge of the population.


3. Define mean, median, and mode. Explain why these measures of central
tendency are important.
  - Mean:
The mean is the arithmetic average of a set of data, calculated by dividing the sum of all observations by the number of observations.

Median:
The median is the middle value of a data set when the observations are arranged in ascending or descending order.

Mode:
The mode is the value that occurs most frequently in a data set.

Importance of measures of central tendency:
Mean, median, and mode are important because they summarize large data sets into a single representative value, help in understanding the central behavior of data, allow comparison between different data sets, and support decision-making and statistical analysis.

Kurtosis is a measure of the peakedness or flatness of a distribution.

It indicates whether the data are more peaked (leptokurtic), normal (mesokurtic), or flatter (platykurtic) than a normal distribution.


```python
#5. Implement a Python program to compute the mean, median, and mode of a given list of numbers.
#numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
#(Include your Python code and output in the code box below.)
import statistics

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

mean = sum(numbers) / len(numbers)
median = statistics.median(numbers)
mode = statistics.mode(numbers)

print("Mean =", mean)
print("Median =", median)
print("Mode =", mode)

#6. Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:
#list_x = [10, 20, 30, 40, 50]
#list_y = [15, 25, 35, 45, 60]

import math

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

n = len(list_x)

mean_x = sum(list_x) / n
mean_y = sum(list_y) / n

covariance = sum((list_x[i] - mean_x) * (list_y[i] - mean_y) for i in range(n)) / (n - 1)

std_x = math.sqrt(sum((x - mean_x) ** 2 for x in list_x) / (n - 1))
std_y = math.sqrt(sum((y - mean_y) ** 2 for y in list_y) / (n - 1))

correlation = covariance / (std_x * std_y)

print("Covariance =", covariance)
print("Correlation Coefficient =", correlation)

#7. Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:
#data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

import matplotlib.pyplot as plt

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

plt.boxplot(data, vert=True, patch_artist=True)
plt.title("Boxplot of Data")
plt.ylabel("Values")
plt.show()

data_sorted = sorted(data)
n = len(data_sorted)

Q1 = data_sorted[n//4]
Q3 = data_sorted[(3*n)//4]
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Q1 =", Q1)
print("Q3 =", Q3)
print("IQR =", IQR)
print("Lower Bound =", lower_bound)
print("Upper Bound =", upper_bound)
print("Outliers =", outliers)
```

Summary Statistics

Mean – shows the average satisfaction.

Median – shows the middle value, useful if data is skewed.

Mode – most frequent score, indicates common sentiment.

Standard Deviation (SD) – measures how much scores vary from the mean (spread of data).

Range – difference between max and min scores.

Visualizations

Histogram – shows frequency of scores, helps see the distribution shape (skewness).

Boxplot – identifies outliers and spread.

Bar chart (optional) – can show frequency of each score for categorical survey data.
